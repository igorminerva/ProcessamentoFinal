```python
import matplotlib.pyplot as plt
import seaborn as sns

# Perform log-transformation
data_log = data.copy()
columns_to_transform = ['Value Added [M.EUR]', 'Employment [1000 p.]', 'GHG emissions [kg CO2 eq.]', 'Energy Carrier Net Total [TJ]']
for column in columns_to_transform:
    data_log[column] = np.log1p(data_log[column])

# Visualize the log-transformed data
plt.figure(figsize=(20, 10))
for i, column in enumerate(columns_to_transform, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data_log[column], kde=True)
    plt.title(f'Log-Transformed {column}')
plt.tight_layout()
plt.show()

# Display basic statistics of the log-transformed data
data_log_stats = data_log[columns_to_transform].describe()

data_log_stats
```

```python
# Calculate the correlation matrix
correlation_matrix = data_log[columns_to_transform].corr()

# Visualize the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Log-Transformed Data')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


features = data_log.drop(columns=['region', 'sector', 'GHG emissions [kg CO2 eq.]'])
target = data_log['GHG emissions [kg CO2 eq.]']


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
# Treine o modelo
rf_regressor.fit(X_train, y_train)
# Faça previsões
y_pred_test = rf_regressor.predict(X_test)
y_pred_train = rf_regressor.predict(X_train)

train_mse = mean_squared_error(y_train, y_pred_train)
test_mse = mean_squared_error(y_test, y_pred_test)

train_mse, test_mse
```

```python
importances = rf_regressor.feature_importances_
features_names = features.columns

# Ordenando as características pela importância
indices = np.argsort(importances)[::-1]

# Visualizando a importância das características
plt.figure(figsize=(10, 6))
plt.title('Importância das Características - Random Forest (Breast Cancer)')
plt.bar(range(features.shape[1]), importances[indices], align='center')
plt.xticks(range(features.shape[1]), np.array(features_names)[indices], rotation=90)
plt.xlim([-1, features.shape[1]])
plt.ylabel('Importância')
plt.xlabel('Características')
plt.show()
```

```python
feature_importance_df = pd.DataFrame({'Feature': features_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

feature_importance_df
```
